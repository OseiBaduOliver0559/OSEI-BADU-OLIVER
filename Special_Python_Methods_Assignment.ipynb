# SPECIAL PYTHON METHODS (DUNDER / MAGIC METHODS)
----
Special methods in Python are surrounded by double underscores (e.g. `__init__`, `__str__`).
They allow custom classes to behave like built-in Python types.

# Understanding Dunder Methods for Object-Oriented Programming

Magic methods (also called **dunder methods** because they have **double underscores** before and after their names) are special methods that allow custom Python classes to interact with Python’s built-in syntax, functions, and operators.

They are automatically invoked by Python when specific operations are performed on objects.

# What Are Magic Methods?
	•	Methods that begin and end with double underscores.
	•	They define how objects respond to built-in operations.
	•	They are invoked automatically by Python.
	•	They support operator overloading.
	•	They make custom classes more intuitive and powerful.

# For example
when you write `len(obj)`, Python internally calls `obj.__len__()`. When you write `obj1 == obj2`, Python calls `obj1.__eq__(obj2)`.

## Categories of Magic Methods

| Object Creation | `__init__()` |
| String Representation | `__str__()`, `__repr__()` |

### Brief Explanation of Each Category

- **Object Creation**: Methods involved in creating and initializing objects.
- **String Representation**: Methods that control how objects are displayed as strings.
- **Comparison**: Methods used to compare objects.
- **Arithmetic Operations**: Methods that enable operator overloading for mathematical operations.
- **Container Operations**: Methods that allow objects to behave like lists, tuples, or dictionaries.
- **Callable Objects**: Methods that allow an object to be called like a function.
- **Iteration**: Methods that make objects iterable using loops.
- **Boolean Evaluation**: Methods that determine the truth value of an object.
- **Object Destruction**: Methods that perform cleanup when an object is destroyed.

## 1. `__init__` — The Constructor Method
This method is automatically called when a new object is created.
It is used to **initialize** the attributes of the object.

```python
class Student:
    def __init__(self, name, age, grade):
        self.name = name
        self.age = age
        self.grade = grade

s1 = Student('Kenneth', 20, 'A')
print('Name:', s1.name)
print('Age:', s1.age)
print('Grade:', s1.grade)
```

---
## 2. `__str__` and `__repr__` — String Representation
`__str__` defines what is displayed when you **print** an object.
`__repr__` is for **debugging** and returns an unambiguous string.

```python
class Student:
    def __init__(self, name, age, grade):
        self.name = name
        self.age = age
        self.grade = grade

    def __str__(self):
        return f'Student Name: {self.name}, Age: {self.age}, Grade: {self.grade}'

    def __repr__(self):
        return f"Student('{self.name}', {self.age}, '{self.grade}')"

s1 = Student('Kenneth', 20, 'A')
print(str(s1))
print(repr(s1))
```

---
## 3. `__len__` — Length of an Object
This method is called when you use `len()` on an object.
It should return an integer representing the size or count.

```python
class Classroom:
    def __init__(self, students):
        self.students = students

    def __len__(self):
        return len(self.students)

c = Classroom(['Kenneth', 'Ama', 'Kofi', 'Abena', 'Kwame'])
print('Number of students:', len(c))
```

---
## 4. `__eq__` — Equality Comparison
This method is called when you use `==` to compare two objects.

```python
class Student:
    def __init__(self, name, score):
        self.name = name
        self.score = score

    def __eq__(self, other):
        return self.score == other.score

s1 = Student('Kenneth', 85)
s2 = Student('Ama', 85)
s3 = Student('Kofi', 70)
print('s1 == s2:', s1 == s2)
print('s1 == s3:', s1 == s3)
```

---
## 5. Operator Overloading — `__add__`, `__sub__`, `__mul__`
These methods define the behaviour of `+`, `-`, and `*` operators on your objects.

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __sub__(self, other):
        return Vector(self.x - other.x, self.y - other.y)

    def __mul__(self, scalar):
        return Vector(self.x * scalar, self.y * scalar)

    def __str__(self):
        return f'Vector({self.x}, {self.y})'

v1 = Vector(4, 6)
v2 = Vector(1, 2)
print('Addition:', v1 + v2)
print('Subtraction:', v1 - v2)
print('Multiplication:', v1 * 3)
```

---
## 6. `__getitem__` — Item Access
This method is called when you access an item using `obj[index]`.

```python
class StudentRegister:
    def __init__(self, students):
        self.students = students

    def __getitem__(self, index):
        return self.students[index]

reg = StudentRegister(['Kenneth', 'Ama', 'Kofi'])
print('First student:', reg[0])
print('Second student:', reg[1])
```

---
## 7. `__setitem__` — Item Assignment
This method is called when you assign a value using `obj[index] = value`.

```python
class StudentRegister:
    def __init__(self):
        self.students = {}

    def __setitem__(self, index, name):
        self.students[index] = name

    def __getitem__(self, index):
        return self.students.get(index, 'Not found')

reg = StudentRegister()
reg[1] = 'Kenneth'
reg[2] = 'Ama'
print('Index 1:', reg[1])
print('Index 2:', reg[2])
```

---
## 8. `__contains__` — Membership Testing
This method is called when you use the `in` keyword to check if an item exists.

```python
class Classroom:
    def __init__(self, students):
        self.students = students

    def __contains__(self, name):
        return name in self.students

c = Classroom(['Kenneth', 'Ama', 'Kofi', 'Abena'])
print('Is Kenneth in class?', 'Kenneth' in c)
print('Is Kwame in class?', 'Kwame' in c)
```

---
## 9. `__call__` — Callable Objects
This method allows an object to be called like a function using `()`.

```python
class Greeter:
    def __init__(self, greeting):
        self.greeting = greeting

    def __call__(self, name):
        return f'{self.greeting}, {name}!'

greet = Greeter('Hello')
print(greet('Kenneth'))
print(greet('Ama'))
```

---
## 10. `__iter__` and `__next__` — Iteration Support
These methods allow an object to be used in a **for loop**.

```python
class CountUp:
    def __init__(self, start, end):
        self.current = start
        self.end = end

    def __iter__(self):
        return self

    def __next__(self):
        if self.current > self.end:
            raise StopIteration
        value = self.current
        self.current += 1
        return value

counter = CountUp(1, 5)
for num in counter:
    print(num)
```

---
## 11. `__bool__` — Boolean Evaluation
This method is called when an object is used in a boolean context like `if obj:` or `bool(obj)`.
It should return `True` or `False`.

```python
class Student:
    def __init__(self, name, score):
        self.name = name
        self.score = score

    def __bool__(self):
        # A student is considered passing if score >= 50
        return self.score >= 50

    def __str__(self):
        return f'{self.name} (Score: {self.score})'

s1 = Student('Kenneth', 85)
s2 = Student('Ama', 40)

if s1:
    print(f'{s1} is PASSING')
else:
    print(f'{s1} is FAILING')

if s2:
    print(f'{s2} is PASSING')
else:
    print(f'{s2} is FAILING')

print('bool(s1):', bool(s1))
print('bool(s2):', bool(s2))
```

---
## 12. `__del__` — Destructor
This method is called when an object is **deleted** or garbage-collected.
Useful for cleanup operations.

```python
class Student:
    def __init__(self, name):
        self.name = name
        print(f'{self.name} record has been created.')

    def __del__(self):
        print(f'{self.name} record has been deleted.')

s1 = Student('Kenneth')
del s1
```
